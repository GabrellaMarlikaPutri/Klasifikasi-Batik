```python
#Import library yang akan digunakan

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# Meload data yang sudah di proses sebelumnya

data = pd.read_csv('DataBatik.csv')
val = pd.read_csv('DataValidasiBatik.csv')
```

```python
# Menginisialisasi Feature dan target yang akan dimodelkan
feature = data.iloc[:, 1:6].values  
target = data.iloc[:, 7].values
feature_val = val.iloc[:, 1:6].values
target_val = val.iloc[:, 7].values
```

```python
def main():
    iteration = input("Input Jumlah Iterasi Yang Diinginkan\t: ")
    epochs = input("Input Jumlah Epoch dalam Satu Iterasi\t: ")


    # Memulai proses pemodelan, kemudian dilanjutkan testing dan validasi
    for iterasi in range(int(iteration)):
        print("============================================================")
        print("\t\t\tIterasi ke-"+str(iterasi+1))
        print("============================================================")
        X_trainDT, X_testDT, y_trainDT, y_testDT = train_test_split(feature, target, test_size=0.3, random_state=11)

        for epoch in range(int(epochs)):
            classifierDT = DecisionTreeClassifier(criterion="gini")
            classifierDT.fit(X_trainDT, y_trainDT)
            prediksiTrain = classifierDT.predict(X_trainDT)
            prediksiTest = classifierDT.predict(X_testDT)

            accsTest = (np.sum(prediksiTest == y_testDT) / float(y_testDT.size)) * 100

            print("Epoch ke-"+str(epoch+1)+"\nTraining Model\n-------> Selesai"+"\n\tTesting\n\t-------> Selesai"
                  +"\tAkurasi: "+str(accsTest)+"%\n")

        prediksiValidasi = classifierDT.predict(feature_val)
        accsVal = (np.sum(prediksiValidasi == target_val) / float(target_val.size)) * 100
        print(".............................................................")
        print("Validasi\n-------> Selesai"+"\t\tAkurasi: "+str(accsVal)+"%\n")
    print("----------------PROSES KLASIFIKASI SELESAI-------------------")
```

```python
main()
```
