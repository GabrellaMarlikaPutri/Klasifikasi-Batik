```python
#Import library yang akan digunakan

import pandas as pd
from time import time
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report, precision_recall_fscore_support
```

```python
# Meload data yang sudah di proses sebelumnya tanpa mengubah dimensi

data_ori = pd.read_csv('DataBatik.csv')
val_ori = pd.read_csv('DataValidasiBatik.csv')
```

```python
data_ori
```

```python
val_ori
```

```python
# Menginisialisasi Feature dan target yang akan dimodelkan
feature_ori = data_ori.iloc[:, 1:6].values  
target_ori = data_ori.iloc[:, 7].values
feature_val_ori = val_ori.iloc[:, 1:6].values
target_val_ori = val_ori.iloc[:, 7].values
```

```python
def pandas_classification_report(y_true, y_pred):
    metrics_summary = precision_recall_fscore_support(y_true=y_true, y_pred=y_pred)

    avg = list(precision_recall_fscore_support(y_true=y_true, y_pred=y_pred, average='weighted'))

    metrics_sum_index = ['precision', 'recall', 'f1-score', 'support']

    class_report_df = pd.DataFrame(list(metrics_summary),index=metrics_sum_index)
    support = class_report_df.loc['support']
    total = support.sum() 
    avg[-1] = total
        
    # class_report_df = pd.DataFrame(index=metrics_sum_index)
    class_report_df['avg / total'] = avg

    return class_report_df.T
```

```python
def main():
    iteration = input("Input Jumlah Iterasi Yang Diinginkan\t: ")
    epochs = input("Input Jumlah Epoch dalam Satu Iterasi\t: ")

    start_awal = time()
    # Memulai proses pemodelan, kemudian dilanjutkan testing dan validasi
    for iterasi in range(int(iteration)):
        print("============================================================")
        print("\t\t\tIterasi ke-"+str(iterasi+1))
        print("============================================================")
       
        for epoch in range(int(epochs)):
            start = time()
            X_trainDT_ori, X_testDT_ori, y_trainDT_ori, y_testDT_ori = train_test_split(feature_ori, target_ori, test_size=0.3, random_state=11)

            classifierDT_ori = DecisionTreeClassifier(criterion="gini")
            classifierDT_ori.fit(X_trainDT_ori, y_trainDT_ori)
            prediksiTrain_ori = classifierDT_ori.predict(X_trainDT_ori)
            print("Epoch ke-"+str(epoch+1)+"\nTraining Model\n-------> Selesai")
            prediksiTest_ori = classifierDT_ori.predict(X_testDT_ori)

            accsTest_ori = (np.sum(prediksiTest_ori == y_testDT_ori) / float(y_testDT_ori.size)) * 100

            print("\tTesting\n\t-------> Selesai"+"\tAkurasi: "+str(accsTest_ori)+"%\n")

        prediksiValidasi_ori = classifierDT_ori.predict(feature_val_ori)
        accsVal_ori = (np.sum(prediksiValidasi_ori == target_val_ori) / float(target_val_ori.size)) * 100
        df_class_report = pandas_classification_report(target_val_ori, prediksiValidasi_ori)
        df_class_report.to_csv("Report-Gambar-Ori"+'_'+str(epochs)+'_'+str(iteration)+'.csv', mode='a', header=False,  sep=';')
        total_val_time = time() - start
        print(".............................................................")
        print("Validasi\n-------> Selesai"+"\t\tAkurasi: "+str(accsVal_ori)+"%\n")
        print("\t\t\t\tWaktu: "+str(total_val_time)+" s")
    print("----------------PROSES KLASIFIKASI SELESAI-------------------")
    total_time = time() - start_awal
    print("Waktu proses keseluruhan: "+str(total_time)+" sekon")
```
